```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://www.koreabaseball.com/Record/History/Top/Hitter.aspx")
req.text
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
print(soup)
```

```python
soup.find("tbody")
```

```python
list_soup = soup.find("tbody").find_all("tr")
```

```python
list_soup[0]
```

```python
list_soup[0].text
```

```python
for i in list_soup :
  wow = i.text
  print(wow)
```

```python
list_soup_2 = []
for tr in list_soup :
  list_soup_2.append(tr.text)
print(list_soup_2)
```
